```python
#setup
import numpy as np
import matplotlib.pyplot as plt
```

```python
#ploting x values in the range -10 to 10 using the formate np.linespace(start, end, num)
x=np.linspace(-10, 10, 100)
#showing the plots

def g(x):
  return x**2+2*x+5
plt.plot(x,g(x))
plt.title('Basic Quadratic Question')
plt.xlabel('x')
plt.ylabel('g(x)')
plt.grid(True)
plt.show()

def h(x):
    return np.sin(x)
plt.plot(x,h(x))
plt.title('Sine Function')
plt.xlabel('x')
plt.ylabel('h(x)')
plt.grid(True)
plt.show()

def i(x):
    return np.sin(x**2+2*x+5)
plt.plot(x,i(x))
plt.title('Combine Quadratic and sine')
plt.xlabel('x')
plt.ylabel('i(x)')
plt.grid(True)
plt.show()

def j(x):
    return np.exp(-0.1*x)
plt.plot(x,j(x))   
plt.title('Exponential Decay')
plt.xlabel('x')
plt.ylabel('j(x)')
plt.grid(True)
plt.show()

def f(x):
    return i(x) + j(x)
plt.plot(x,f(x))    
plt.title('Complete Function')
plt.xlabel('x')
plt.ylabel('f(x)')
plt.grid(True)
plt.show()
    
           

    
    
    
```
