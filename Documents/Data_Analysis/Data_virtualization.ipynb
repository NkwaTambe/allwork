```python
# Get the count of each transaction type
transaction_type_counts = df['type'].value_counts()
print(transaction_type_counts)
print()
# Create a bar chart of the transaction type counts
plt.bar(transaction_type_counts.index, transaction_type_counts.values)
plt.xlabel('Transaction Type')
plt.ylabel('Count')
plt.title('Transaction Type Distribution')
plt.show()
```

```python
# Create a histogram of the transaction amounts
plt.hist(df['amount'], bins=10)
plt.xlabel('Transaction Amount')
plt.ylabel('Frequency')
plt.title('Transaction Amount Distribution')
plt.show()
```

```python
#Using a pie chart, visualize the proportion of fraudulent vs. non-fraudulent transaction
fraud_count = ['isFraud'].value_counts()
plt.pie(fraud_count, labels=fraud_count.index, autopct='%1.2f%%')
plt.title('proportion of fraudulent vs non-proportion of fraudulent')
plt.axis('equal')
plt.show()
```

```python
# Plot a scatter plot with oldbalanceorg on the x-axis and newbalanceOrig on the y-axis
plt.figure(figsize=(10, 6))
plt.scatter(df['oldbalanceOrg'], df['newbalanceOrig'], alpha=0.5)
plt.xlabel('Old Balance')
plt.ylabel('New Balance')
plt.title('Old Balance vs. New Balance')
plt.show()
```
